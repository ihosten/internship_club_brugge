## Testing
The final output has to be filtered for position and club/competition.
For this it is explored how the optimal data format can be attained.

```python
# To avoid testing on the entire data frame a few standard metrics/columns are stored in a list
test_metrics = ['normalised Distance','normalised Running Distance','normalised HSR Distance',
               'normalised Sprinting Distance', 'PSV-99']
```

```python
grouped1 = (
    df.select(["season_name", "competition_name", "position"] + test_metrics)
    .group_by(["season_name", "competition_name", "position"])
    .agg(
        [pl.col(col).mean().alias(f"{col}_mean") for col in test_metrics] +
        [pl.col(col).std().alias(f"{col}_std") for col in test_metrics] + 
        [pl.col(col).median().alias(f"{col}_median") for col in test_metrics]+ 
        [pl.col(col).quantile(0.1).alias(f"{col}_10%_percentile") for col in test_metrics]+ 
        [pl.col(col).quantile(0.2).alias(f"{col}_20%_percentile") for col in test_metrics]+ 
        [pl.col(col).quantile(0.3).alias(f"{col}_30%_percentile") for col in test_metrics]+ 
        [pl.col(col).quantile(0.4).alias(f"{col}_40%_percentile") for col in test_metrics]+ 
        [pl.col(col).quantile(0.5).alias(f"{col}_50%_percentile") for col in test_metrics]+ 
        [pl.col(col).quantile(0.6).alias(f"{col}_60%_percentile") for col in test_metrics]+ 
        [pl.col(col).quantile(0.7).alias(f"{col}_70%_percentile") for col in test_metrics]+ 
        [pl.col(col).quantile(0.8).alias(f"{col}_80%_percentile") for col in test_metrics]+ 
        [pl.col(col).quantile(0.9).alias(f"{col}_90%_percentile") for col in test_metrics]+ 
        [pl.col(col).quantile(1.0).alias(f"{col}_100%_percentile") for col in test_metrics]
    )
)
grouped1.sort("season_name")
```

```python
# define the path to the benchmark excel
wb = "match_benchmarks.xlsx"
```

```python
# store all the unique postition to loop through
positions = grouped1.select('position').unique('position')
positions
```

```python
# store all the unique competition years to loop through
season_names = grouped1.select('season_name').unique('season_name').to_series().to_list()

# store all the unique postition to loop through
# if needed list can be manually created to loop in a set order
positions = grouped1.select('position').unique('position').to_series().to_list()
#print(positions)

# store all the unique competitions to loop through
# if needed list can be manually created to loop in a set order
competitions = grouped1.select('competition_name').unique('competition_name').to_series().to_list()



metric_notations = ['_mean','_std','_median','_10%_percentile','_20%_percentile',
                   '_30%_percentile','_40%_percentile','_50%_percentile','_60%_percentile',
                   '_70%_percentile','_80%_percentile','_90%_percentile','_100%_percentile']

test_metrics = ['normalised Distance','normalised Running Distance','normalised HSR Distance',
               'normalised Sprinting Distance', 'PSV-99']

header = test_metrics

# create a counter for testing
counter = 0

for season in tqdm(season_names):

    for position in tqdm(positions):
        if counter >= 1:
            break
        counter += 1

        # filter df on position
        filtered_position = grouped1.filter(pl.col("position") == position)

        for competition in competitions:
            # filter on competition
            filtered_competition = filtered_position.filter(pl.col("competition_name") == competition)  
            #check the filterd df
            #print(filtered_competition)

            #initialize a list to store benchmark values
            benchmarks = {}

            for metric in test_metrics:
                met_scores = []

                for notation in metric_notations:
                    # select the needed dataframe
                    met_scores.append(filtered_competition.select(f'{metric}{notation}'))
                
                benchmarks[metric] = met_scores

            # create the df to write to the excell
            df2write  = pl.from_dict(benchmarks)

            # write the stats to the excell and the correct sheet    

            worksheet_name = f"{position}_{season}"

            # Replace invalid Excel characters with underscores
            worksheet_name = re.sub(r'[\\/*?:\[\]]', '_', worksheet_name)
            
            df2write.write_excel(
                workbook=wb,
                worksheet= worksheet_name
                )



        # Leave a few blank rows before entering the next competition 
        #  
```
