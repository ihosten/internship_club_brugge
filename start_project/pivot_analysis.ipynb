```python
#create a subset of the df to test the pivot function on a few columns

df_subset =  df.select(['normalised Distance','normalised Running Distance','normalised HSR Distance',
               'normalised Sprinting Distance', 'PSV-99', "season_name", "competition_name", "position"])
df_subset.head()
```

```python
#expr over
# met Expr zo std en quantiles
# eerqst bewerd$ken daarna pivot

# test of the pivot function
value_cols = ['normalised Distance','normalised Running Distance',
                            'normalised HSR Distance','normalised Sprinting Distance', 
                            'PSV-99']
expressions = ["mean", 
               "median",
               #"pl.col().std()"
             ]
pivot_results = []

for expression in expressions:
    pivot = df_subset.pivot(
        on = "position", 
        index = ["season_name","competition_name"], 
        values = [
            'normalised Distance','normalised Running Distance',
            'normalised HSR Distance','normalised Sprinting Distance', 
            'PSV-99'
        ],
        aggregate_function = expression
        ).with_columns([
            pl.lit(expression).alias("statistical_measure")
        ])
    pivot_results.append(pivot)

results = pl.concat(pivot_results)
```

```python
# testing if the aggregation can be done for seperate positions
value_cols = ['normalised Distance','normalised Running Distance',
                            'normalised HSR Distance','normalised Sprinting Distance', 
                            'PSV-99']
expressions = ["mean", 
               "median",
             ]
pivot_results = []

counter = 0

for position in tqdm(positions):
    if counter >= 1:
        break
    counter += 1

    filtered_position = df_subset.filter(pl.col("position") == position)
    
    for expression in expressions:
        pivot = filtered_position.pivot(
            on = "position", 
            index = ["season_name","competition_name"], 
            values = value_cols,
            aggregate_function = expression
            ).with_columns([
                pl.lit(expression).alias("statistical_measure")
            ])
        pivot_results.append(pivot)

    results = pl.concat(pivot_results)
```
